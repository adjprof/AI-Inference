```
from sklearn.cluster import KMeans
from google.colab import files
import pandas as pd
import io

from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```
## Import Data
# Use categories: state,zipcode,ethnicity,income,age,sex,education,marriage,childen,party
# Data goes into 'X', labels go into 'y'
uploaded = files.upload()


# raw data
#X = pd.read_csv(io.BytesIO(uploaded['IA_synthetic_full_data_Output_Oct21.csv']))

# anonymized data
X = pd.read_csv(io.BytesIO(uploaded['k_3_anon_IA_data_full_Oct21.csv']))

# Fix some things
X = X.drop('state', axis=1)
X = X.drop([8, 186254], axis=0)
```

```
## Split data into train, dev, test; 70% train, 15% dev, 15% test
X_train, X_test = train_test_split(X, test_size=0.3, random_state=1)
X_dev, X_test = train_test_split(X_test, test_size=0.5, random_state=1)
```

```
# Debuggg!!!
#X = X.drop([8, 186254], axis=0)
X = X.drop('party', axis=1)
```

```
# Make Classifier
kmeans = KMeans(n_clusters=3)

# Fit
kmeans.fit(X_train)

# Predict labels (?)
y_kmeans = kmeans.predict(X_dev)

# What are those?
print(kmeans.labels_)
print(y_kmeans)
```
